```ruby
require 'terminal-table'
require 'yaml'

    table = Terminal::Table.new :headings => ['File', 'Key', 'Value']

    regexp_str = []
    regexp_str << '*.yml'     # search *.yml
    regexp_str << 'group_vars/*.yml' # search */*.yml e.g. group_vars, host_vars
    regexp_str << 'host_vars/*.yml' # search */*.yml e.g. group_vars, host_vars
    regexp_str << '**/defaults/*' # search roles/*/vars/*.yml
    regexp_str << '**/vars/*' # search roles/*/vars/*.yml
    # regexp_str << '**{,/*/**}/*.{yml}'
    regexp_str.each do |str|
      Dir.glob(str) do |f|
        next unless FileTest.file?(f) # skip directory
       
        yml = YAML.load_file(f)
      
        if yml.nil?
          puts "No Variables in #{f}"
          puts "\n"
          next
        end
        
        p f
        puts "-----------------------------------------------"
          
        if str == '*.yml'
          yml.each do |h|
            next unless h.instance_of?(Hash) && h.has_key?('vars')

            h['vars'].each do |key, value|
                puts "the string was '*.yml'"
                puts key
                puts value
              table << [f, key, value]
            end
          end
        else
          yml.each do |key, value|
                  
                if value.class == Hash
                    table = Terminal::Table.new do |t|
                      t.title = "#{key} Information"
                      t.headings = ["Key", "Value"]                      
                      value.each do |k,v|
                        t.add_row([k, v])
                      end
                    end
                  puts table
                    next
                end
                if value.class == Array
                    table = Terminal::Table.new do |t|
                      t.title = "#{key} Information"
                      t.headings = ["Key", "Value"]   
                    value.each do |va|
                        va.each do |k,v|
                            t.add_row([k,v])
                        end
                      end
                    end
                  puts table
                  next
                end
                    
                puts "#{key}: #{value}"
                puts "\n"
            table << [f, key, value]
          end
        end

        puts "\n\n"
      end
    end

```
